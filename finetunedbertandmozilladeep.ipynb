```python
# Install required libraries
!pip install transformers datasets
!pip install torchaudio
!pip install librosa

# Load the model and processor from Hugging Face
from transformers import Wav2Vec2Processor, Wav2Vec2ForCTC
import torch
import librosa

model_name = "Baghdad99/saad-speech-recognition-hausa-audio-to-text"
processor = Wav2Vec2Processor.from_pretrained(model_name)
model = Wav2Vec2ForCTC.from_pretrained(model_name)

# Load the audio file
audio_path = "/content/testn.ogg"
speech_array, sampling_rate = librosa.load(audio_path, sr=16000)

# Preprocess the audio
input_values = processor(speech_array, sampling_rate=16000, return_tensors="pt").input_values

# Perform inference
with torch.no_grad():
    logits = model(input_values).logits

# Get the predicted ids
predicted_ids = torch.argmax(logits, dim=-1)

# Decode the ids to text
transcription = processor.batch_decode(predicted_ids)[0]
print("Transcription:", transcription)
```

```python
import time

# Install required libraries
!pip install transformers
!pip install librosa

# Load the model and processor from Hugging Face
from transformers import WhisperProcessor, WhisperForConditionalGeneration
import torch
import librosa

model_name = "Baghdad99/saad-speech-recognition-hausa-audio-to-text"
processor = WhisperProcessor.from_pretrained(model_name)
model = WhisperForConditionalGeneration.from_pretrained(model_name)

# Load the audio file
audio_path = "/content/testn.ogg"
speech_array, sampling_rate = librosa.load(audio_path, sr=16000)

# Preprocess the audio
input_features = processor(speech_array, sampling_rate=16000, return_tensors="pt").input_features

# Measure the time taken for transcription
start_time = time.time()

# Perform inference
with torch.no_grad():
    predicted_ids = model.generate(input_features)

end_time = time.time()

# Decode the ids to text
transcription = processor.batch_decode(predicted_ids, skip_special_tokens=True)[0]
transcription_time = end_time - start_time

print("Transcription:", transcription)
print(f"Time taken for transcription: {transcription_time:.2f} seconds")
```
